```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

# Task 1: Replace Hardcoded Prompts

```python

def explain_topic(topic):
    return f"Explain {topic} in simple terms for beginners"

print(explain_topic("Artificial Intelligence"))

# LangChain PromptTemplate version
prompt_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

output = prompt_template.format(topic="Artificial Intelligence")
print(output)
```

This task converts a hardcoded prompt into a reusable LangChain PromptTemplate. The template can be used with different topics without changing the code.


# Task 2: Multi-Input Prompt System

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

This task demonstrates how PromptTemplate can handle multiple dynamic inputs, making prompts more flexible and reusable.


# Task 3: Prompt Variations Engine

```python
# Teaching template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling template
storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print("Teaching Prompt:")
print(teaching_template.format(topic=topic))

print("\nInterview Prompt:")
print(interview_template.format(topic=topic))

print("\nStorytelling Prompt:")
print(storytelling_template.format(topic=topic))
```

# Task 4: ChatPromptTemplate System

```python
roles = {
    "teacher": "You are a knowledgeable teacher who explains concepts clearly.",
    "interviewer": "You are an interviewer who asks challenging questions.",
    "motivator": "You are a motivator who encourages and inspires learners."
}

def create_chat_prompt(role, topic):
    chat_template = ChatPromptTemplate.from_messages([
        ("system", roles[role]),
        ("human", "Explain {topic}")
    ])
    return chat_template.format_messages(topic=topic)

messages = create_chat_prompt("teacher", "Neural Networks")

for message in messages:
    print(f"{message.type.upper()}: {message.content}")
```

# Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        print("Invalid audience! Assigning default: beginner")
        audience = "beginner"

    if tone not in valid_tones:
        print("Invalid tone! Assigning default: formal")
        tone = "formal"

    return audience, tone

print(validate_inputs("beginner", "fun"))
print(validate_inputs("kids", "friendly"))
```

# Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} style",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

# Task 7: Template Reusability Test

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "students", "tone": "fun"},
    {"topic": "SQL", "audience": "analysts", "tone": "formal"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"},
    {"topic": "Power BI", "audience": "professionals", "tone": "casual"}
]

for i, data in enumerate(inputs, 1):
    print(f"Output {i}:")
    print(reusable_template.format(**data))
    print()
```
